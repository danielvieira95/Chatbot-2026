```python
from google.colab import userdata
chat_api = userdata.get('api_chat26')
```

```python
#chat_api
```

```python
from llama_index.llms.groq import Groq # importando o metodo Groq a partir do framework llama
```

```python
llm = Groq(model='llama-3.3-70b-versatile',api_key=chat_api)
response = llm.complete('Qual é a substancia do aroma do alecrim')
```

```python
response
```

```python
paragrafos = response.text.split("\n\n")
for paragrafos in paragrafos:
  print(paragrafos)
print()
```
