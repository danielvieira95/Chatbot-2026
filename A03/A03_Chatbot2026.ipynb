```python
# importando a biblioteca do google colab
from google.colab import userdata
chat_api = userdata.get('api_chat26') # armazenando a chave da api
```

```python
from groq import Groq # importando a biblioteca Groq
```

```python
client = Groq(api_key=chat_api) # cria um cliente do Groq para realizar as
# requisições para a api
# Faz a solicitação para o modelo
completion =client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {
          "role":"system","content":"Você é um assistente de agricultor"
        },
        {
            "role":"user","content":"Qual a semente é indicada para \n o solo arenoso em maio ?"
        }
    ],
    temperature =0,
    max_tokens=1024,
    top_p=1,
    stream = False,

)
#Exibe a resposta gerada

print(completion.choices[0].message.content)
```

```python
# Solicitando para que o usuario do chat digite sua pergunta
user_prompt = input('Digite sua pergunta:')
client = Groq(api_key=chat_api) # cria um cliente do Groq para realizar as
# requisições para a api
# Faz a solicitação para o modelo
completion =client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {
          "role":"system","content":"Você é um assistente de agricultor"
        },
        {
            "role":"user","content":user_prompt
        }
    ],
    temperature =0.5,
    max_tokens=1024,
    top_p=1,
    stream = False,

)
#Exibe a resposta gerada
print("\n Resposta do assistente: \n ")
print(completion.choices[0].message.content)
```

```python
client = Groq(api_key=chat_api) # cria um cliente do Groq para realizar as
while True:

  # Solicitando para que o usuario do chat digite sua pergunta
  user_prompt = input('\n Digite sua pergunta sobre agricultura ou digite 15 para sair:')


  # requisições para a api
  # Faz a solicitação para o modelo
  completion =client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {
          "role":"system","content":"Você é um assistente de agricultor"
        },
        {
            "role":"user","content":user_prompt
        }
      ],
      temperature =0.5,
      max_tokens=1024,
      top_p=1,
      stream = False,
)
      #Exibe a resposta gerada
  print("\n Resposta do assistente: \n ")
  print(completion.choices[0].message.content)
  if user_prompt.strip()=='15':
    print( 'Encerrando o chatbot')
    break
```
